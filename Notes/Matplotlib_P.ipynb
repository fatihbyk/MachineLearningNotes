```python
import numpy as np
x=np.arange(0,10)
y=x**2
```

```python
plt.plot(x, y, 'r') # 'r' is the color red
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('String Title Here')
```

```python
fig=plt.figure()
#plt.subplot(2,2,1)
x1=fig.add_subplot(222)#ilk sayı satırı,2. sayı sütunu, 3 sayı kaçıncı grafik olduğnunu gösterir
x2=fig.add_subplot(221)
x2.set_title('ikinci')
```

```python
y1=plt.subplot(2,2,1)
y2=plt.subplot(2,2,2)
y2.set_title('fatih')
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("default axes ranges")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("tight axes")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5]) #limit belirleme
axes[2].set_title("custom axes range");
axes[2].set_xticklabels(['a','sadd','sad','s','asasdd','asdasdd'])
```

```python
fig = plt.figure()  # an empty figure with no Axes
fig, ax = plt.subplots()  # a figure with a single Axes
fig, axs = plt.subplots(2, 2)  # a figure with a 2x2 grid of Axes
```

# OO ile  grafik oluşturma


```python
fig=plt.figure()
axes=fig.add_axes([0.1, 0.1, 0.8, 0.8])
#Burada yazdığım ilk değer soldan ne kadar içerde başlayacağı,ikinci değer alttan ne kadar yüksekte olacağı,
#üçüncü değer ne kadar uzunlukta olması gerektiği,dördüncü değer ise ne kadar yükseklikte olması gerektiğini belirtmektedir
axes.plot(x, y, 'r')
axes.set_xlabel('Set X Label') # Notice the use of set_ to begin methods
axes.set_ylabel('Set y Label')
axes.set_title('Set Title')
axes.set_xticks([2,3,4,5,6])
axes.set_xticklabels([2,3,4,5,6])
```

```python
# Creates blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # inset axes  #grafiğin içindeki yer için ayarlama yapıyoruz

# Larger Figure Axes 1
axes1.plot(x, y, 'b')
axes1.set_xlabel('X_label_axes2')
axes1.set_ylabel('Y_label_axes2')
axes1.set_title('Axes 2 Title')

# Insert Figure Axes 2
axes2.plot(y, x, 'r')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title');
```

```python
# Empty canvas of 1 by 2 subplots
fig, axes = plt.subplots(nrows=1, ncols=2)  #axes aslında liste for la kullanılabilir, subplots figure deki sayıları otomatik alıyor
axes[0].plot(x,y)
axes[0].set_title('first')
```

```python
fig, axes = plt.subplots(figsize=(12,3))

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

```python
fig2=plt.figure()
ax=fig2.add_axes([0,1,1,1])
ax.plot(x,x**2,label='x squared')
ax.plot(x,x**3,label='y cubed')
ax.legend(loc=0) #lejant etkin olması için grafiklerde label belirtmek gerekir. loc komutu yer için 0 en iyi çözümü getirir
# 10 center mesela. eğer hiçbiri uymazsa 
```

# Plot Appreance


```python
fig=plt.figure()
ax=fig.add_axes([1,1,1,1])
ax.plot(x,y,color='red') #RGB HEX CODE
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='red',linewidth=2,alpha=0.5,linestyle='-',marker='o',markersize=10,
       markerfacecolor='green',markeredgewidth=3,markeredgecolor='yellow')    #alpha transparency
```
