```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x=np.linspace(0,10,100)
```

```python
plt.plot(x,np.sin(x))
```

```python
fig=plt.figure()
plt.subplot(2,1,2)
plt.plot(x,np.sin(x),label='Sin')
plt.xlabel('Sin x')

plt.plot(x,np.cos(x),linewidth=5,label='Cos')
plt.ylabel('Cos x')
plt.legend()
```

```python
import seaborn as sns
import pandas as pd
```

```python
iris=sns.load_dataset('iris') #internette popüler olan datasetlerini çekmeye yarıyor
iris.head()
```

```python
iris['sepal_length'].plot()
```

```python
sns.lineplot(data=iris,y='petal_length',x=iris.index,hue='species') # hue kategorik olan değişkenleri ayırıp bize gösteriyor
```

```python
fig,ax=plt.subplots()
ax.scatter(iris['sepal_length'],iris ['sepal_width'])
ax.set_title('iris')
ax.set_xlabel('sepal length')
ax.set_ylabel('sepal width')
```

```python
iris.groupby(by='species').mean()['sepal_length'] #sadece group by ile bıraklılırsa oluşturuluyor ekrana bastırmıyor
```

```python
data=iris.groupby(by='species').mean()['sepal_length'] 
ax=sns.barplot(x=data.index,y=data.values)
ax.set_title('sepal length')
ax.set_ylabel('Max')
```

```python
#histogram

sns.histplot(data=iris,x='sepal_width',bins=30)
#yatayda da görülebilir hue .ok mantıklı değil burada bins aralıkla bölmemizi sağlar
```

```python
iris.corr()
sns.heatmap(iris.corr(),annot=True) #annot okumayı kolaylaştırıyor
```

```python
sns.boxplot(data=iris,y='sepal_length',x='species')
```
