```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

# Distribution Plots

```python
sns.displot(tips['total_bill'],kde=False,bins=30)  #kde false line ı ortadan kaldırır
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter',color='r',) #kind defaultu scatter
```

```python
sns.pairplot(data=tips,hue='sex',palette='coolwarm') # hue kategorik veriye giriyor yani non numeric olması gerekir
```

```python
sns.rugplot(tips['total_bill'],y=tips['tip'],hue=tips['sex'])
```

# Categorical Plots

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std,seed=True)  #estimator defaultu meandir
```

```python
sns.countplot(x='sex',data=tips) # x kategorik data alıyor tek parametre
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)  #Jitter False olursa direkt çzigi şekinde verir
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,dodge=False,color='black') #dodge farkı ??
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='violin') # hepsi çağrılıyor ama tek tek yapma daha iyi
```

# Matrix Plots

```python
flights=sns.load_dataset('flights')
```

```python
tc=tips.corr()
```

```python
sns.heatmap(tc,annot=True,robust=True) #vmin,vmax correlation değerliernin sınırlarını belirtiyor. robust?
```

```python
my_colors=[(0.2,0.3,0.3),(0.4,0.5,0.4),(0.1,0.7,0),(0.1,0.7,0)]
sns.heatmap(tc,annot=True,vmin=0.4,vmax=0.9,robust=False,cmap=my_colors,linewidth=10,linecolor='w',square=True)
#cmap->corr map için renk paleti ya liste şeklinde alır ya da ismi string değer ile alır
```

```python
sns.set( rc={'font.family':'Arial','figure.figsize':(150,5),})
#sns.set_theme() ile içine ne konulacağı bakılabilir
sns.heatmap(tc,annot=True,vmin=0.4,vmax=0.9,robust=False,cmap=my_colors,linewidth=10,linecolor='w',square=True)
```

```python
fp=flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.set(rc={'figure.figsize':(10,8)})
sns.heatmap(fp,linecolor='w',lw=2)
```

```python
sns.clustermap(fp,cmap='coolwarm')  #method biraz daha ML ile alakalı
```

# Grids


```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
g=sns.PairGrid(iris)
#g.map(plt.scatter)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
g=sns.FacetGrid(data=tips,col='time',row='smoker') #kategorik ya da discrete
sns.set(style='darkgrid')
g.map(sns.distplot,'total_bill')
```

# Regression Plots


```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',aspect=3,scatter_kws={'s':100},sharex=True) #aspect boyutu belirler
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time',aspect=1,height=4) #hue yerine col ya da row parametresi alabilir
```

```python
sns.set_style('ticks')  #whitegrid  ,darkgrid,ticks
sns.countplot(x='sex',data=tips)
sns.despine(left=True,bottom=True)   # default olarak sağ ve üst truedur bakılabilir
```
